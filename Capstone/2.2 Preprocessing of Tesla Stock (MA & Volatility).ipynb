```python
tsla.columns=[x.lower() for x in  tsla.columns]
```

**Create Moving Averages 5,10,20, 50 Days**

```python
tsla['MA5']=tsla.close.rolling(5, center=True).mean().shift()
tsla['MA10']=tsla.close.rolling(10, center=True).mean().shift()
tsla['MA20']=tsla.close.rolling(20, center=True).mean().shift()
tsla['MA50']=tsla.close.rolling(50, center=True).mean().shift()
```

**Compute Log Change (Instantaneous Rate of Return)**

```python
tsla['change']=np.log(tsla['close']/tsla['close'].shift())
#returns natural logarithm (inverse exponential function)
```

**Compute Rolling Historical Volatility, offset using shift() method**

```python
tsla['volatility']=tsla.change.rolling(21).std().shift()
```
